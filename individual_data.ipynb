```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_selection import SelectFromModel

results = []
feature_sets = []

for i in range(1, 16):

    #df = pd.read_csv(f"triple/data{i}.csv")
    df = pd.read_csv(f"/content/drive/MyDrive/ORNL_Project/triple/data{i}.csv")

    X = df.drop(columns=['marker'])
    y = df['marker']

    # Handle inf/nan
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X.fillna(X.median(), inplace=True)

    # =========================
    # FEATURE SELECTION
    # =========================
    rf_selector = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        class_weight='balanced'
    )

    rf_selector.fit(X, y)

    selector = SelectFromModel(
        rf_selector,
        max_features=40,
        threshold=-np.inf
    )

    selector.fit(X, y)

    X_selected = pd.DataFrame(
        selector.transform(X),
        columns=X.columns[selector.get_support()]
    )

    selected_features = X.columns[selector.get_support()].tolist()
    feature_sets.append(selected_features)

    # =========================
    # TRAIN / TEST SPLIT
    # =========================
    X_train, X_test, y_train, y_test = train_test_split(
        X_selected,
        y,
        test_size=0.2,
        random_state=42,
        stratify=y
    )

    # =========================
    # RANDOM FOREST
    # =========================
    rf = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        class_weight='balanced'
    )

    rf.fit(X_train, y_train)
    rf_pred = rf.predict(X_test)

    rf_acc = accuracy_score(y_test, rf_pred)
    rf_report = classification_report(y_test, rf_pred, output_dict=True)

    # =========================
    # DECISION TREE
    # =========================
    dt = DecisionTreeClassifier(
        random_state=42,
        class_weight='balanced'
    )

    dt.fit(X_train, y_train)
    dt_pred = dt.predict(X_test)

    dt_acc = accuracy_score(y_test, dt_pred)
    dt_report = classification_report(y_test, dt_pred, output_dict=True)

    # =========================
    # STORE RESULTS
    # =========================
    results.append({
        'Dataset': f'Dataset-{i}',
        'Classifier': 'RF',
        'Accuracy': round(rf_acc, 4),
        'Precision': round(rf_report['macro avg']['precision'], 4),
        'Recall': round(rf_report['macro avg']['recall'], 4),
        'F1-Score': round(rf_report['macro avg']['f1-score'], 4)
    })

    results.append({
        'Dataset': '',
        'Classifier': 'DT',
        'Accuracy': round(dt_acc, 4),
        'Precision': round(dt_report['macro avg']['precision'], 4),
        'Recall': round(dt_report['macro avg']['recall'], 4),
        'F1-Score': round(dt_report['macro avg']['f1-score'], 4)
    })

    print("\n")

# =========================
# FINAL TABLE
# =========================
results_df = pd.DataFrame(results)

print("\n" + "="*80)
print("TABLE 6: Per-Dataset Performance")
print("="*80)

print(results_df.to_string(index=False))
```
